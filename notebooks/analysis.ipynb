```python
df_clean=df.copy()
```

```python
df_clean['timestamp'] = pd.to_datetime(df_clean['timestamp'])
df_clean['date'] = pd.to_datetime(df_clean['date'])
```

```python
text_columns = [
    'receiver_type',
    'day_of_week',
    'transaction_type',
    'payment_app',
    'device_type',
    'status',
    'user_city_tier',
    'user_kyc_status'
]

for column in text_columns:
    df_clean[column] = df_clean[column].str.strip()
```

```python
df_clean['time_since_last_txn_missing'] = df_clean['time_since_last_txn_min'].isnull().astype(int)
df_clean['transaction_velocity_missing'] = df_clean['transaction_velocity'].isnull().astype(int)
df_clean['amount_deviation_score_missing'] = df_clean['amount_deviation_score'].isnull().astype(int)
```

```python
negative_time_count = (df_clean['time_since_last_txn_min'] < 0).sum()

negative_time_count
```

```python
df_clean['time_since_last_txn_negative_flag'] = (df_clean['time_since_last_txn_min'] < 0).astype(int)
```

```python
df_clean.loc[df_clean['time_since_last_txn_min'] < 0, 'time_since_last_txn_min'] = np.nan
```

```python
time_since_median = df_clean['time_since_last_txn_min'].median()

df_clean['time_since_last_txn_min'] = df_clean['time_since_last_txn_min'].fillna(time_since_median)
```

```python
transaction_velocity_median = df_clean['transaction_velocity'].median()

df_clean['transaction_velocity'] = df_clean['transaction_velocity'].fillna(transaction_velocity_median)
```

```python
amount_deviation_median = df_clean['amount_deviation_score'].median()

df_clean['amount_deviation_score'] = df_clean['amount_deviation_score'].fillna(amount_deviation_median)
```

```python
#Data transformation
df_clean['month'] = df_clean['timestamp'].dt.month
df_clean['month_name'] = df_clean['timestamp'].dt.month_name()
```

```python
df_clean['quarter'] = df_clean['timestamp'].dt.quarter
```

```python
def get_transaction_period(hour):
    if 6 <= hour <= 11:
        return 'Morning'
    elif 12 <= hour <= 16:
        return 'Afternoon'
    elif 17 <= hour <= 20:
        return 'Evening'
    elif 21 <= hour <= 23:
        return 'Late Night'
    else:
        return 'Night'
    
df_clean['transaction_period'] = df_clean['hour_of_day'].apply(get_transaction_period)    
```

```python
def get_amount_bucket(amount):
    if amount < 250:
        return 'Low'
    elif 250 <= amount < 1000:
        return 'Medium'
    elif 1000 <= amount < 5000:
        return 'High'
    else:
        return 'Very High'

df_clean['amount_bucket'] = df_clean['amount'].apply(get_amount_bucket)
```

```python
df_clean['is_successful'] = (df_clean['status'] == 'Success').astype(int)
```

```python
df_clean['is_successful'].mean()*100
```

```python
df_clean['is_failed'] = (df_clean['status'] == 'Failed').astype(int)
```

```python
df_clean['kyc_risk_flag'] = (df_clean['user_kyc_status'] == 'Not Verified').astype(int)
```

```python
amount_deviation_threshold = df_clean['amount_deviation_score'].quantile(0.95)

df_clean['high_amount_deviation_flag'] = (
    df_clean['amount_deviation_score'] >= amount_deviation_threshold
).astype(int)
```

```python
velocity_threshold = df_clean['transaction_velocity'].quantile(0.95)

df_clean['high_velocity_flag'] = (
    df_clean['transaction_velocity'] >= velocity_threshold
).astype(int)
```

```python
df_clean['high_failed_attempts_flag'] = (
    df_clean['failed_attempts_last_24h'] >= 3
).astype(int)
```

```python
df_clean['risk_score'] = (
    df_clean['kyc_risk_flag']
    + df_clean['new_device_flag']
    + df_clean['ip_location_mismatch']
    + df_clean['high_failed_attempts_flag']
    + df_clean['high_velocity_flag']
    + df_clean['high_amount_deviation_flag']
    + df_clean['is_night_transaction']
)
```

```python
def get_risk_level(score):
    if score <=1:
        return 'Low Risk'
    elif score <=3:
        return 'Medium Risk'
    else:
        return 'High Risk'
df_clean['risk_level'] = df_clean['risk_score'].apply(get_risk_level)    
```

```python
new_columns = [
    'month',
    'month_name',
    'quarter',
    'transaction_period',
    'amount_bucket',
    'is_successful',
    'is_failed',
    'kyc_risk_flag',
    'high_amount_deviation_flag',
    'high_velocity_flag',
    'high_failed_attempts_flag',
    'risk_score',
    'risk_level'
]

df_clean[new_columns].head()
```

```python
#Analysis Reports
total_transactions = df_clean.shape[0]
total_spending = df_clean['amount'].sum()
average_transaction = df_clean['amount'].mean()

print(f"Total Transactions: {total_transactions}")
print(f"Total Spending/Volume: ₹{total_spending:,.2f}")
print(f"Average Transaction Value: ₹{average_transaction:,.2f}")
```

```python
monthly_summary = df_clean.groupby('month_name').agg(
    total_transactions =('transaction_id','count'),
    total_amount=('amount','sum')
).sort_values('total_amount',ascending=False)

monthly_summary
```

```python
app_summary = df_clean.groupby('payment_app').agg(
    transaction_count=('transaction_id', 'count'),
    total_volume=('amount', 'sum')
).sort_values('transaction_count', ascending=False)

app_summary
```

```python
fraud_by_app = df_clean.groupby('payment_app').agg(
    total_transactions=('transaction_id', 'count'),
    fraud_cases=('is_fraud', 'sum')
)

fraud_by_app['fraud_rate_percentage'] = (fraud_by_app['fraud_cases'] / fraud_by_app['total_transactions']) * 100
fraud_by_app = fraud_by_app.sort_values('fraud_rate_percentage', ascending=False)

fraud_by_app
```

```python
risk_summary = df_clean.groupby('risk_level').agg(
    transaction_count=('transaction_id', 'count'),
    total_amount_at_risk=('amount', 'sum')
).sort_values('transaction_count', ascending=False)

risk_summary
```

```python
kyc_fraud = df_clean.groupby('kyc_risk_flag').agg(
    total_txns=('transaction_id', 'count'),
    fraud_cases=('is_fraud', 'sum')
)

kyc_fraud['fraud_rate_%'] = (kyc_fraud['fraud_cases'] / kyc_fraud['total_txns']) * 100
kyc_fraud = kyc_fraud.sort_values('fraud_rate_%', ascending=False)

kyc_fraud
```

```python
time_fraud = df_clean.groupby('transaction_period').agg(
    total_txns=('transaction_id', 'count'),
    fraud_cases=('is_fraud', 'sum')
)

time_fraud['fraud_rate_%'] = (time_fraud['fraud_cases'] / time_fraud['total_txns']) * 100
time_fraud = time_fraud.sort_values('fraud_rate_%', ascending=False)

time_fraud
```

```python
amount_fraud = df_clean.groupby('amount_bucket').agg(
    total_txns=('transaction_id', 'count'),
    fraud_cases=('is_fraud', 'sum')
)

amount_fraud['fraud_rate_%'] = (amount_fraud['fraud_cases'] / amount_fraud['total_txns']) * 100
amount_fraud = amount_fraud.sort_values('fraud_rate_%', ascending=False)

amount_fraud
```

```python
df_clean.to_csv('../outputs/cleaned_transactions.csv',index=False)

monthly_summary.to_csv('../outputs/monthly_summary.csv', index=True)
app_summary.to_csv('../outputs/payment_app_summary.csv', index=True)
time_fraud.to_csv('../outputs/time_fraud_analysis.csv', index=True)
amount_fraud.to_csv('../outputs/amount_fraud_analysis.csv', index=True)
kyc_fraud.to_csv('../outputs/kyc_fraud_analysis.csv', index=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 

sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10,6)
```

```python
time_data = df_clean.groupby('transaction_period').agg(
    total_txns=('transaction_id', 'count'),
    fraud_cases=('is_fraud', 'sum')
)
time_data['fraud_rate'] = (time_data['fraud_cases'] / time_data['total_txns']) * 100
time_data = time_data.reset_index().sort_values('fraud_rate', ascending=False)

plt.figure(figsize=(10, 6))

sns.barplot(x='transaction_period', y='fraud_rate', data=time_data,hue='transaction_period', palette='magma',legend=False)

plt.title('Fraud Risk by Time of Day', fontsize=16, fontweight='bold')
plt.xlabel('Time of Day', fontsize=12)
plt.ylabel('Fraud Rate (%)', fontsize=12)

plt.savefig('../charts/fraud_by_time.png', bbox_inches='tight')
plt.show()
```

```python
kyc_data = df_clean.groupby('kyc_risk_flag').agg(
    total_txns=('transaction_id', 'count'),
    fraud_cases=('is_fraud', 'sum')
)
kyc_data['fraud_rate'] = (kyc_data['fraud_cases'] / kyc_data['total_txns']) * 100
kyc_data = kyc_data.reset_index()

kyc_data['KYC Status'] = kyc_data['kyc_risk_flag'].map({0: 'Verified (Low Risk)', 1: 'Not Verified (High Risk)'})

plt.figure(figsize=(8, 5))

sns.barplot(x='KYC Status', y='fraud_rate', data=kyc_data, 
            hue='KYC Status', palette='Set2', legend=False)

plt.title('Impact of KYC Verification on Fraud Rate', fontsize=16, fontweight='bold')
plt.ylabel('Fraud Rate (%)', fontsize=12)
plt.xlabel('')

plt.savefig('../charts/kyc_fraud.png', bbox_inches='tight')
plt.show()
```

```python
app_data = df_clean.groupby('payment_app').agg(
    total_txns=('transaction_id', 'count'),
    fraud_cases=('is_fraud', 'sum')
)
app_data['fraud_rate'] = (app_data['fraud_cases'] / app_data['total_txns']) * 100
app_data = app_data.reset_index().sort_values('fraud_rate', ascending=False)

plt.figure(figsize=(10, 6))

sns.barplot(x='payment_app', y='fraud_rate', data=app_data, 
            hue='payment_app', palette='viridis', legend=False)

plt.title('Fraud Rate by Payment Application', fontsize=16, fontweight='bold')
plt.xlabel('Payment Application', fontsize=12)
plt.ylabel('Fraud Rate (%)', fontsize=12)

plt.xticks(rotation=45) 

plt.tight_layout()
plt.savefig('../charts/fraud_by_app.png', bbox_inches='tight')
plt.show()
```
